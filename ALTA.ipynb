```python
path = "/content/devtestset.txt"
path1 = "/content/dev_filename_embedding1.txt"
import numpy as np

list_emb = []
with open(path, "r") as f:
  with open(path1,"r") as f1:
    text = f.readlines()
    text1 = f1.readlines()
    for line in text:
      sum = [0.0,0.0,0.0]
      new_line=line.strip("\n").split(" ")
      print(new_line,len(new_line)-2)
      for i in range(len(new_line)-2):
        file = new_line[i+2]
        f1.seek(0)
        for line1 in text1:
          new_line1 = line1.strip("\n").split("\t")
          if(new_line1[0]==file):
            #print(new_line1[1].shape)
            list_str = new_line1[1].replace("   "," ").replace("  "," ").replace("[","").replace(" ]","]").replace("]","").split(' ')
            #print(list_str)
            list2 = []
            for j in list_str:
              list2.append(float(j))
            arr1 = np.array(list2)
            print(arr1)
            sum+=arr1
            break;
      print("sum：", sum/(len(new_line)-2))
      list_emb.append(list(sum/(len(new_line)-2)))
```
